# Object Oriented Method (OOM)

```python
fig, ax = plt.subplots(2, 1)

ax[0].plot(age, salary, "r")
ax[1].plot(age, salary_2, "b")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2)

ax[0].plot(age, salary, "r")
ax[1].plot(age, salary_2, "b")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2,3)  # 2 satir 3 sutun

ax[0][0].plot(age, salary, "r")
ax[0][1].plot(age, salary_2, "b")
ax[0][2].plot(age, salary, "r")

ax[1][0].plot(age, salary_2, "b")
ax[1][1].plot(age, salary, "r")
ax[1][2].plot(age, salary_2, "b")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(3,2)

ax[0][0].plot(age, salary, "r")
ax[0][1].plot(age, salary_2, "b")

ax[1][0].plot(age, salary_2, "b")
ax[1][1].plot(age, salary, "r")

ax[2][0].plot(age, salary_2, "b")
ax[2][1].plot(age, salary, "r")

plt.tight_layout()
plt.show()
```

```python
x = np.arange(0, 10)

fig, ax = plt.subplots(2, 2, figsize = (12, 6))

plt.suptitle("LineStyles & Markers & Markersize & Color",fontsize=16, y = 0.95)

ax[0,0].plot(x, "b", marker='+',markersize=5)
ax[0,1].plot(x-2, "y", marker='o',ms=5) #ms can be used for markersize
ax[1,0].plot(x-3, "black", marker='s',ms=5, lw=0) # make linewidth zero to see only markers
ax[1,1].plot(x-4, "g", marker='1',ms=5)

plt.show()
```

# Multiple Plots

```python
x = np.arange(-50, 50)
y1 = np.power(x, 2)
y2 = np.power(x, 3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)



fig1, ax1 = plt.subplots(nrows = 3, ncols = 3, figsize= (20, 20))

ax1[0, 0].plot(x, y1, "tab:blue") # set the color of line chart
ax1[0, 0].set_title("Square Function") # setting title of subplot
ax1[0, 0].set_xlabel("X", fontsize = 18)
ax1[0, 0].set_ylabel("Y", fontsize = 18)

ax1[0, 1].plot(x, y2, "tab:orange") # set the color of line chart
ax1[0, 1].set_title("Cubic Function") # setting title of subplot
ax1[0, 1].set_xlabel("X", fontsize = 18)
ax1[0, 1].set_ylabel("Y", fontsize = 18)

ax1[0, 2].plot(x, y3, "tab:green") # set the color of line chart
ax1[0, 2].set_title("Sine Function") # setting title of subplot
ax1[0, 2].set_xlabel("X", fontsize = 18)
ax1[0, 2].set_ylabel("Y", fontsize = 18)

ax1[1, 0].plot(x, y4, "r-") # set the color of line chart
ax1[1, 0].set_title("Cosine Function") # setting title of subplot
ax1[1, 0].set_xlabel("X", fontsize = 18)
ax1[1, 0].set_ylabel("Y", fontsize = 18)

ax1[1, 1].plot(x, y5, "b-") # set the color of line chart
ax1[1, 1].set_title("Tangent Function") # setting title of subplot
ax1[1, 1].set_xlabel("X", fontsize = 18)
ax1[1, 1].set_ylabel("Y", fontsize = 18)

ax1[1, 2].plot(x, y6, "g-") # set the color of line chart
ax1[1, 2].set_title("Hyperbolic Tangent Function") # setting title of subplot
ax1[1, 2].set_xlabel("X", fontsize = 18)
ax1[1, 2].set_ylabel("Y", fontsize = 18)

ax1[2, 0].plot(x, y7, "m-") # set the color of line chart
ax1[2, 0].set_title("Hyperbolic Sine Function") # setting title of subplot
ax1[2, 0].set_xlabel("X", fontsize = 18)
ax1[2, 0].set_ylabel("Y", fontsize = 18)

ax1[2, 1].plot(x, y8, "y*") # set the color of line chart
ax1[2, 1].set_title("Hyperbolic Cosine Function") # setting title of subplot
ax1[2, 1].set_xlabel("X", fontsize = 18)
ax1[2, 1].set_ylabel("Y", fontsize = 18)

ax1[2, 2].plot(x, y9, "kv") # set the color of line chart
ax1[2, 2].set_title("Exp Function") # setting title of subplot
ax1[2, 2].set_xlabel("X", fontsize = 18)
ax1[2, 2].set_ylabel("Y", fontsize = 18)

plt.show()
```
