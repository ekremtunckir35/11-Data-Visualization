Kutu GrafiÄŸinin Ã–ÄŸeleri

Kutunun alt ve Ã¼st kenarlarÄ±, verinin 1. Ã§eyreÄŸi (Q1) ve 3. Ã§eyreÄŸini (Q3) temsil eder.<br>

Kutunun ve bÄ±yÄ±klarÄ±n dÄ±ÅŸÄ±nda kalan noktalar, aykÄ±rÄ± deÄŸerleri temsil eder.<br>
