Kodun Detaylı Açıklaması
sns.displot():

Seaborn'un dağılım grafiği (distribution plot) fonksiyonudur. Veri kümesinin belirli bir değişkeninin dağılımını görselleştirmek için kullanılır.
data=cars[cars.cylinders.isin([4, 6, 8])]:

data parametresi, grafik için kullanılacak veri kümesini belirtir. Burada, cars veri kümesinden yalnızca cylinders değeri 4, 6 veya 8 olan satırlar seçilmektedir. cars.cylinders.isin([4, 6, 8]) ifadesi bu filtrelemeyi yapar.
x="weight":

x parametresi, x ekseninde hangi değişkenin gösterileceğini belirtir. Burada, arabaların ağırlıkları (weight) x ekseninde gösterilecektir.
col="origin":

col parametresi, grafiğin sütunlara nasıl ayrılacağını belirtir. Burada, arabaların menşeine göre sütunlara ayrılmıştır (origin değişkeni).
row="cylinders":

row parametresi, grafiğin satırlara nasıl ayrılacağını belirtir. Burada, arabaların silindir sayılarına göre satırlara ayrılmıştır (cylinders değişkeni).
hue="cylinders":

hue parametresi, grafikte renklerle hangi değişkenin ayrıştırılacağını belirtir. Burada, silindir sayısına göre renklendirme yapılmaktadır (cylinders değişkeni).
palette="plasma":

palette parametresi, grafik için kullanılacak renk paletini belirtir. Burada, "plasma" renk paleti seçilmiştir. Seaborn ve Matplotlib'te çeşitli renk paletleri bulunmaktadır ve "plasma" bunlardan biridir.
Görsel Çıktı
Bu kod, arabaların ağırlık dağılımını aşağıdaki şekilde görselleştirecektir:

Her bir sütun, arabaların üretildiği ülkeyi (menşei) temsil eder.
Her bir satır, arabaların silindir sayılarını temsil eder.
Grafikteki her bir renk, farklı silindir sayılarını temsil eder.
X ekseni arabaların ağırlıklarını, y ekseni ise bu ağırlık değerlerinin dağılımını gösterir.
Örnek Grafiğin Görünümü
Örneğin, üç sütun ve üç satırdan oluşan bir grafik elde edersiniz:

sns.countplot():

Seaborn'un count plot fonksiyonudur. Bu fonksiyon, belirli bir kategorik değişkenin her bir kategori için kaç tane gözlemin olduğunu görselleştirmek için kullanılır.
data=tips:

data parametresi, grafik için kullanılacak veri kümesini belirtir. Burada tips veri kümesi kullanılmıştır. tips, genellikle bir restoranın müşterilerinden alınan bahşişlerin ve çeşitli değişkenlerin (gün, toplam hesap, bahşiş miktarı, cinsiyet vb.) yer aldığı örnek bir veri kümesidir.
x="day":

x parametresi, x ekseninde hangi değişkenin gösterileceğini belirtir. Burada day değişkeni x ekseninde gösterilecektir. day, müşterilerin hangi günlerde restoranı ziyaret ettiğini belirten kategorik bir değişkendir (örneğin, Thur, Fri, Sat, Sun).
Görsel Çıktı
Bu kodun çıktısı, tips veri kümesindeki day değişkeninin her bir kategorisi (gün) için kaç tane gözlem olduğunu gösteren bir sütun grafiği olacaktır. Grafikte:

X ekseni, günleri (örneğin, Perşembe, Cuma, Cumartesi, Pazar) temsil eder.
Y ekseni, her bir gün için gözlemlenen müşteri sayısını (frekanstı) temsil eder.
Her bir sütun, belirli bir günde kaç müşteri olduğunu görsel olarak gösterir.
Örnek Grafiğin Görünümü
Örneğin, aşağıdaki gibi bir grafik elde edersiniz:

X ekseninde Thur, Fri, Sat, Sun gibi günler.
Y ekseninde her bir gün için gözlemlenen müşteri sayıları.
Perşembe günü 10 müşteri, Cuma günü 15 müşteri, Cumartesi günü 20 müşteri ve Pazar günü 25 müşteri varsa, her bir gün için bu sayıları gösteren sütunlar olur.
Bu grafik, müşterilerin hangi günlerde daha sık restoranı ziyaret ettiklerini görsel olarak anlamanızı sağlar. Bu tür bilgiler, restoran işletmecileri için hangi günlerde daha yoğun olduklarını belirlemek açısından faydalı olabilir.

```python
fig, ax = plt.subplots()

ax = sns.countplot(data = tips, x = "day")

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+0.2))
```

#### Kodun Detaylı Açıklaması

**fig, ax = plt.subplots():**

**Matplotlib kütüphanesi ile bir figür (fig) ve eksen (ax) oluşturur. Bu, grafiğin çizileceği alanı hazırlar.<br>**

**ax = sns.countplot(data=tips, x="day"):<br>**

**Seaborn kullanarak tips veri kümesindeki day değişkenine göre bir sayım grafiği oluşturur ve bu grafiği ax ekseni üzerinde çizer.**

**data=tips parametresi, grafiğin tips veri kümesi kullanılarak çizileceğini belirtir.**

**x="day" parametresi, x ekseninde day değişkeninin gösterileceğini belirtir.**

**for p in ax.patches:**

**Bu döngü, grafikteki her bir sütunu (patches) iterasyonla dolaşır. ax.patches, grafikteki tüm sütunları (dikdörtgenleri) içerir.**

**ax.annotate((p.get_height()), (p.get_x() + 0.3, p.get_height() + 0.2)):**

**Bu satır, her bir sütunun üstüne yüksekliğini yazan bir anotasyon (etiket) ekler.**

**p.get_height(), sütunun yüksekliğini (y ekseni değeri) alır.**

**(p.get_x() + 0.3, p.get_height() + 0.2), etiketin yerleştirileceği koordinatları belirtir. p.get_x() sütunun x eksenindeki başlangıç noktasıdır. Buna 0.3 ekleyerek etiketi sütunun ortasına doğru kaydırır. p.get_height() + 0.2 etiketi sütunun üstüne biraz mesafeli yerleştirir.**

#### Görsel Çıktı

#### Bu kodun çıktısı, her bir gün için müşteri sayısını gösteren bir sayım grafiği olacaktır. Ayrıca, her sütunun üstünde ilgili müşteri sayısını belirten etiketler (sayısal değerler) olacaktır.

#### Kodun Detaylı Açıklaması 

sns.boxplot():

Seaborn'un box plot fonksiyonudur. Bu fonksiyon, verinin dağılımını beş sayı özetine (min, 1. çeyrek, medyan, 3. çeyrek, max) göre görselleştirir ve aykırı değerleri (outliers) gösterir.
x="day":

x parametresi, x ekseninde hangi değişkenin gösterileceğini belirtir. Burada day değişkeni (gün) x ekseninde gösterilecektir.

##### Grafiğin Açıklaması
Grafikte, tips veri kümesindeki bahşiş miktarlarının (tip) haftanın dört günü (Thur, Fri, Sat, Sun) için dağılımı gösterilmektedir.

Kutu Grafiğinin Öğeleri
Kutu (Box):<br>
Kutunun alt ve üst kenarları, verinin 1. çeyreği (Q1) ve 3. çeyreğini (Q3) temsil eder.<br>
Kutunun içindeki çizgi, verinin medyanını (Q2) gösterir.<br>
Bıyıklar (Whiskers):<br>
Bıyıklar, verinin minimum ve maksimum değerlerine kadar uzanır, ancak genellikle 1.5 IQR (interquartile range) sınırları içinde kalırlar.<br>
Aykırı Değerler (Outliers):<br>
Kutunun ve bıyıkların dışında kalan noktalar, aykırı değerleri temsil eder.<br>

## Comparison Plot(Karsilastirma)

### Joint Plot

```python
sns.jointplot(x = "total_bill", y = "tip", data = tips, kind = "scatter", alpha = 0.7);
```

Bu grafik, Seaborn kütüphanesi kullanılarak tips veri kümesindeki total_bill (toplam hesap) ve tip (bahşiş) değişkenleri arasındaki ilişkiyi görselleştiren bir jointplot (ortak çizim grafiği) olarak oluşturulmuştur. Bu tür grafikler, iki değişkenin dağılımını ve ilişkisini tek bir grafik üzerinde gösterir.

### Kodun Detaylı Açıklaması  

sns.jointplot():

Seaborn'un jointplot fonksiyonudur. Bu fonksiyon, iki değişkenin birlikte dağılımını ve ilişkisini gösteren bir grafik oluşturur.
x="total_bill":

x parametresi, x ekseninde hangi değişkenin gösterileceğini belirtir. Burada total_bill (toplam hesap) x ekseninde gösterilecektir.
y="tip":

y parametresi, y ekseninde hangi değişkenin gösterileceğini belirtir. Burada tip (bahşiş) y ekseninde gösterilecektir.
data=tips:

data parametresi, grafik için kullanılacak veri kümesini belirtir. Burada tips veri kümesi kullanılmıştır.
kind="scatter":

kind parametresi, grafiğin türünü belirtir. Burada, bir saçılma grafiği (scatter plot) seçilmiştir. Diğer olası türler reg (regresyon), hex (hexbin), kde (kernel density estimate) olabilir.
alpha=0.7:

alpha parametresi, nokta işaretlerinin saydamlığını belirtir. 0.7 değeri, %70 opaklık anlamına gelir ve bu, noktaların biraz saydam olmasını sağlar, böylece yoğun bölgeler daha net görülebilir.<br>

#### Grafiğin Açıklaması

Ana Grafik (Orta Kısım):
Ana grafik, total_bill ve tip değişkenleri arasındaki ilişkiyi scatter plot (dağılım grafiği) olarak gösterir.
Her bir nokta, bir müşteri hesabını ve verilen bahşiş miktarını temsil eder.

Bu tür grafikler, çok değişkenli verilerin incelenmesinde ve değişkenler arasındaki ilişkilerin keşfedilmesinde oldukça faydalıdır.


sns.pairplot():
Seaborn'un pair plot fonksiyonudur. Bu fonksiyon, veri kümesindeki tüm sayısal değişkenlerin çiftleri arasındaki ilişkileri scatter plot ve histogramlar kullanarak görselleştirir.

penguins:

Grafik için kullanılacak veri kümesini belirtir. Burada penguins veri kümesi kullanılmıştır. penguins veri kümesi, penguen türleri ve özelliklerini içeren bir veri kümesidir.<br>

hue="species":<br>

palette="viridis":

palette parametresi, grafikte kullanılacak renk paletini belirtir. Burada "viridis" renk paleti seçilmiştir. Viridis, güzel ve kolay ayrıştırılabilir renkler sunan bir palettir.

diag_kind="hist":<br>

diag_kind parametresi, diyagonal (köşegen) elemanların türünü belirtir. Burada histogram (hist) seçilmiştir. Bu, her bir değişkenin kendi dağılımını histogram olarak gösterir.<br>


corner=True:<br>

corner parametresi, grafiğin alt üçgenini (veya köşegen altını) çizer. True olarak ayarlandığında, yalnızca alt üçgen ve köşegen çizilir, böylece grafikteki bilgi tekrarı azalır ve daha sade bir görünüm elde edilir.

#### Scatter Plots (Saçılma Grafikleri):

Histogramlar:<br>

Köşegen Altı:<br>

corner=True parametresi nedeniyle, grafik yalnızca alt üçgeni ve köşegen elemanları içerir. Bu, veri tekrarını azaltır ve grafiği daha okunabilir kılar.<br>

Dağılımlar:

### Facet Grid

```python
g = sns.FacetGrid(data = tips, col = "time", row = "smoker")
g = g.map(plt.hist, "total_bill");
```

Bu kod, Seaborn kütüphanesi kullanılarak tips veri kümesindeki total_bill (toplam hesap) değişkeninin time (öğle yemeği veya akşam yemeği) ve smoker (sigara içen veya içmeyen) değişkenlerine göre dağılımını görselleştiren bir FacetGrid (yüzey ızgarası) oluşturur ve histogramlar ile bu dağılımları gösterir.<br>

### Kodun Detaylı Açıklaması

sns.FacetGrid():

Seaborn'un FacetGrid fonksiyonudur. Bu fonksiyon, çoklu küçük grafiklerin oluşturulmasına olanak tanır ve veri setinin belirli kategorik değişkenlere göre bölünmüş alt grafiklerini oluşturur.
data=tips:

data parametresi, grafik için kullanılacak veri kümesini belirtir. Burada tips veri kümesi kullanılmıştır. tips, restoran müşterilerinden alınan bahşişlerin ve çeşitli değişkenlerin (gün, toplam hesap, bahşiş miktarı, cinsiyet vb.) yer aldığı örnek bir veri kümesidir.
col="time":

col parametresi, grafiğin sütunlara nasıl ayrılacağını belirtir. Burada time değişkeni (öğle yemeği veya akşam yemeği) kullanılmıştır, yani grafikler time değişkenine göre sütunlara ayrılacaktır.
row="smoker":

row parametresi, grafiğin satırlara nasıl ayrılacağını belirtir. Burada smoker değişkeni (sigara içen veya içmeyen) kullanılmıştır, yani grafikler smoker değişkenine göre satırlara ayrılacaktır.
g.map(plt.hist, "total_bill"):

map fonksiyonu, belirtilen grafiği (burada plt.hist, yani histogram) her bir alt grafikte uygulamak için kullanılır.
plt.hist fonksiyonu, histogram çizmek için Matplotlib kütüphanesi tarafından sağlanır.
"total_bill" parametresi, histogramın hangi değişkene göre çizileceğini belirtir. Burada total_bill değişkenine göre histogramlar çizilecektir.
Grafiğin Açıklaması
Bu kod, total_bill değişkeninin time ve smoker kategorilerine göre dağılımını dört ayrı histogram grafiği olarak görselleştirir:

Sütunlar:

Her sütun time değişkenine göre ayrılmıştır: Lunch (öğle yemeği) ve Dinner (akşam yemeği).
Satırlar:

Her satır smoker değişkenine göre ayrılmıştır: Yes (sigara içen) ve No (sigara içmeyen).
Örnek Çıktı
İlk sütun, öğle yemeği (Lunch) için histogramları gösterir.

İlk satır, sigara içmeyen (No) müşteriler için öğle yemeği hesaplarının dağılımını gösterir.
İkinci satır, sigara içen (Yes) müşteriler için öğle yemeği hesaplarının dağılımını gösterir.
İkinci sütun, akşam yemeği (Dinner) için histogramları gösterir.

İlk satır, sigara içmeyen (No) müşteriler için akşam yemeği hesaplarının dağılımını gösterir.
İkinci satır, sigara içen (Yes) müşteriler için akşam yemeği hesaplarının dağılımını gösterir.
Grafikten Çıkarımlar
Hesap Dağılımları:
Öğle yemeği ve akşam yemeği hesaplarının dağılımlarını sigara içme durumu ile birlikte inceleyebilirsiniz.
Kategorilere Göre Farklılıklar:
Sigara içen ve içmeyen müşterilerin toplam hesap miktarlarında farklılık olup olmadığını gözlemleyebilirsiniz.
Öğle yemeği ve akşam yemeği arasındaki hesap miktarı farklılıklarını görebilirsiniz.
Bu grafik, restoran müşterilerinin toplam hesaplarının farklı zamanlarda ve sigara içme durumlarına göre nasıl değiştiğini analiz etmek için kullanışlıdır.
