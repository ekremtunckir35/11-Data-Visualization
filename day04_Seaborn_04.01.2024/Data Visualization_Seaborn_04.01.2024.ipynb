## Scatter Plot

- Scatter plot dagilim grafikleri arasinda gecen farkli ozellikerin birbiriyle iliskisine bakar.<br>
- Scatter plot, iki değişken arasındaki potansiyel ilişkileri veya korelasyonları görselleştirmek için kullanılır.<br>

```python
sns.scatterplot(data = tips, x = "total_bill", y = "tip")
plt.show()
```

**- Bu grafik, bir scatter plot (dağılım grafiği) olup, toplam hesap tutarı (total_bill) ile verilen bahşiş (tip) arasındaki ilişkiyi göstermektedir. <br>**
**- Scatter plot, iki değişken arasındaki potansiyel ilişkileri veya korelasyonları görselleştirmek için kullanılır.<br>**

**-Grafiğin Açıklaması
X Ekseni (total_bill): Yatay eksen, toplam hesap tutarını temsil eder. Değerler yaklaşık 3.07 ile 50.81 arasında değişmektedir.<br>**

**Y Ekseni (tip): Dikey eksen, verilen bahşiş miktarını temsil eder. Değerler yaklaşık 1.00 ile 10.00 arasında değişmektedir.<br>**

**Veri Noktaları: Her bir mavi nokta, bir gözlemi temsil eder. Yani, her nokta belirli bir toplam hesap tutarı ve buna karşılık gelen bahşiş miktarını gösterir.<br>**

**Grafikten Çıkarımlar<br>**

**Pozitif Korelasyon: Grafik genel olarak pozitif bir eğilim gösteriyor. Bu, toplam hesap tutarının arttıkça verilen bahşiş miktarının da artma eğiliminde olduğunu gösterir.<br>**

**Dağılım ve Yoğunluk:<br>**

**Hesap tutarları ve bahşiş miktarları düşük değerlerde daha yoğunlaşmış durumda. Özellikle, toplam hesap tutarının 10 ile 20 arasında olduğu ve bahşiş miktarının 1 ile 4 arasında yoğunlaştığı görülüyor.<br>**

**Yüksek hesap tutarları (30'dan fazla) için bahşiş miktarlarının daha geniş bir aralıkta dağıldığı, ancak yine de genellikle 2 ile 6 arasında yoğunlaştığı görülüyor.<br>**

**Özet<br>**

**Grafik, toplam hesap tutarı ile bahşiş miktarı arasında genel olarak pozitif bir ilişki olduğunu göstermektedir. Daha yüksek hesap tutarları, genellikle daha yüksek bahşişlerle ilişkilidir. Ancak, belirli bir hesap tutarı aralığında (örneğin, 10-20) bahşiş miktarlarının oldukça değişken olduğu ve çeşitli faktörlere bağlı olarak farklılık gösterebileceği de görülmektedir.<br>**

## Distribution Plot

**Distribution Plot (dağılım grafiği), bir veri setinin dağılımını görselleştirmek için kullanılır. Bu tür grafikler, veri setindeki değerlerin frekansını, yoğunluğunu veya olasılık dağılımını gösterir. Seaborn kütüphanesinde en sık kullanılan dağılım grafikleri distplot, histplot, kdeplot ve rugplot gibi fonksiyonlarla oluşturulabilir.<br>**

**Distribution Plot Türleri**<br>

**Histogram (histplot): Veriyi belirli aralıklara (bin) böler ve her aralıktaki veri noktalarının sayısını (frekansını) gösterir.<br>**

**KDE Plot (kdeplot): Kernel Yoğunluk Tahmini (Kernel Density Estimation), veri noktalarının yoğunluğunu gösterir ve veri dağılımının sürekli bir eğrisini çizer.<br>**

**Rug Plot (rugplot): Veri noktalarının yerlerini belirten kısa çizgilerden oluşur. Genellikle histogram veya KDE plot ile birlikte kullanılır.<br>**

```python
sns.displot(x="total_bill", data = tips);
```

```python
# Histogram oluşturma
sns.histplot(data=tips, x="total_bill", bins=30, kde=False)
plt.xlabel("Toplam Hesap")
plt.ylabel("Frekans")
plt.title("Toplam Hesap Dağılımı")
plt.show()
```

## KDE Plot (kdeplot) Örneği

**KDE plot, veri yoğunluğunu gösterir ve histogramın daha pürüzsüz bir versiyonudur**

```python
sns.kdeplot(data=tips, x="total_bill")
plt.xlabel("Toplam Hesap")
plt.ylabel("Yoğunluk")
plt.title("Toplam Hesap Yoğunluğu")
plt.show()
```

## Histogram ve KDE Plot Birlikte Kullanma

**Histogram ve KDE plot'u bir arada kullanarak, hem frekansı hem de yoğunluğu görselleştirebilirsiniz**

```python
# Histogram ve KDE plot birlikte oluşturma
sns.histplot(data=tips, x="total_bill", bins=30, kde=True)
plt.xlabel("Toplam Hesap")
plt.ylabel("Frekans / Yoğunluk")
plt.title("Toplam Hesap Dağılımı ve Yoğunluğu")
plt.show()
```

## Rug Plot (rugplot) Örneği

**Rug plot, veri noktalarının yerlerini göstermek için kullanılır ve genellikle histogram veya KDE plot ile birlikte kullanılır**

```python
# Histogram ve Rug plot birlikte oluşturma
sns.histplot(data=tips, x="total_bill", bins=30, kde=True)
sns.rugplot(data=tips, x="total_bill")
plt.xlabel("Toplam Hesap")
plt.ylabel("Frekans / Yoğunluk")
plt.title("Toplam Hesap Dağılımı ve Yoğunluğu ile Veri Noktaları")
plt.show()
```

**Özet**
**Histogram: Veriyi belirli aralıklara böler ve her aralıktaki veri sayısını gösterir.<br>**

**KDE Plot: Veri yoğunluğunu gösterir ve veri dağılımının sürekli bir eğrisini çizer.<br>**

**Rug Plot: Veri noktalarının yerlerini belirten kısa çizgilerden oluşur.<br>**

**Birlikte Kullanım: Histogram, KDE ve rug plot bir arada kullanılarak, veri setinin dağılımı hakkında daha kapsamlı bilgi elde edilebilir.<br>**

**Distribution plot'lar, veri dağılımını anlamak ve görselleştirmek için güçlü araçlardır. Veri setinizin yapısını, merkezi eğilimlerini, yayılımını ve olası uç değerlerini (outliers) görselleştirmenize yardımcı olur.**<br>

## Histplot

**Histplot, veri dağılımını görselleştirmek için kullanılan bir histogram oluşturma fonksiyonudur. Histogram, veriyi belirli aralıklara (bin) böler ve her aralıktaki veri noktalarının sayısını (frekansını) gösterir. Bu, veri setinin genel dağılımını anlamak ve belirli aralıklardaki yoğunluğu görmek için kullanışlıdır.<br>

**Histogramın Temel Özellikleri<br>**

**Binler: Histogramda veri aralıkları "bin" olarak adlandırılır. Her bir bin, belirli bir aralıktaki veri noktalarının sayısını temsil eder.<br>**

**Frekans: Y ekseni, her bir binde bulunan veri noktalarının sayısını (frekansını) gösterir.<br>**

**Dağılım: Histogram, veri setinin genel dağılımını ve hangi aralıklarda yoğunlaştığını gösterir.<br>**

***Histplot Kullanımı***<br>

**Seaborn kütüphanesinde histplot fonksiyonu kullanılarak histogram oluşturulabilir. Aşağıda bazı örnekler ve açıklamalar bulunmaktadır.**


**Özet**<br>

**Binler: Histogramdaki aralıklar. Bin sayısı arttıkça detay artar, azaldıkça genelleme yapılır.<br>**

**Frekans: Y ekseninde gösterilen değerler, her binde kaç veri noktası olduğunu gösterir.<br>**

**KDE Eğrisi: Histogramın üzerinde, veri yoğunluğunu daha pürüzsüz bir şekilde gösteren eğri.<br>**

**Hue: Veriyi kategorik bir değişkene göre renklendirerek farklı grupları karşılaştırma imkanı sağlar.<br>**

**Histplot, veri setinin dağılımını anlamak için oldukça kullanışlıdır. Veri setinizin belirli aralıklarda nasıl dağıldığını, hangi aralıkların daha yoğun olduğunu ve olası anormallikleri tespit etmek için kullanabilirsiniz.<br>**

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 5))

sns.rugplot(data=tips, x = "total_bill", ax = ax[0])
sns.histplot(data=tips, x = "tip", kde = True, ax = ax[1]);
```

## KDE Plot (Kernel Densty Estimation)

**KDE Plot (Kernel Density Estimation), veri setinin olasılık yoğunluk fonksiyonunu pürüzsüz bir şekilde tahmin eden ve görselleştiren bir grafiktir.<br>**

**KDE plot, histogramın daha pürüzsüz bir versiyonu olarak düşünülebilir. Histogram gibi verinin dağılımını gösterir, ancak veriyi belirli aralıklara bölmek yerine sürekli bir eğri kullanarak yoğunluğu gösterir. Bu, veri setinin genel yapısını ve yoğunluk bölgelerini daha net bir şekilde görselleştirir.<br>**

**KDE Plot'un Temel Özellikleri**<br>

**Yoğunluk Eğrisi: Veri noktalarının yoğunluğunu sürekli bir eğri ile gösterir.<br>**

**Pürüzsüzlük: KDE, veriyi pürüzsüzleştirir ve belirli aralıklarla sınırlı kalmaz.<br>**

**Kernel: KDE, veri noktalarının yoğunluğunu hesaplamak için kernel fonksiyonları kullanır. En yaygın kullanılan kernel, Gaussian (normal) kerneldir.<br>**

**Bant Genişliği (Bandwidth): KDE eğrisinin pürüzsüzlüğünü belirleyen bir parametredir. Bant genişliği küçükse eğri daha detaylı, büyükse daha genel olur.**<br>

**KDE Plot Kullanımı**<br>

**Seaborn kütüphanesinde kdeplot fonksiyonu kullanılarak KDE plot oluşturulabilir.**<br>


**Özet**<br>

**Yoğunluk Eğrisi: KDE plot, veri noktalarının yoğunluğunu sürekli bir eğri ile gösterir.<br>**

**Pürüzsüzlük: Histogramdan farklı olarak, KDE plot veriyi pürüzsüzleştirir.<br>**

**Kernel: Veri yoğunluğunu hesaplamak için kernel fonksiyonları kullanılır.<br>**

**Bant Genişliği: KDE eğrisinin pürüzsüzlüğünü belirler. Küçük bant genişliği daha detaylı, büyük bant genişliği daha genel bir eğri oluşturur.<br>**

**Hue: Veriyi kategorik bir değişkene göre renklendirerek farklı grupları karşılaştırma imkanı sağlar.<br>**

**KDE plot, veri setinin genel dağılımını ve yoğunluk bölgelerini anlamak için güçlü bir araçtır. Veriyi daha pürüzsüz bir şekilde görselleştirerek, belirli bölgelerdeki yoğunluğu ve olası anormallikleri tespit etmenizi sağlar.<br>**

### Basit KDE Plot

```python
# KDE plot oluşturma
sns.kdeplot(data=tips, x="total_bill")
plt.xlabel("Toplam Hesap")
plt.ylabel("Yoğunluk")
plt.title("Toplam Hesap Yoğunluğu")
plt.show()
```

Bu örnekte:

data=tips: Veri seti olarak tips kullanılıyor.<br>

x="total_bill": KDE plot, total_bill sütununun dağılımını gösterecek.

### Bant Genişliğini Ayarlama

Bant genişliği, KDE eğrisinin pürüzsüzlüğünü kontrol eder. Aşağıdaki örnek, bant genişliğini ayarlayarak KDE plot oluşturur:

```python
# KDE plot oluşturma, bant genişliği ayarlama
sns.kdeplot(data=tips, x="total_bill", bw_adjust=0.5)
plt.xlabel("Toplam Hesap")
plt.ylabel("Yoğunluk")
plt.title("Toplam Hesap Yoğunluğu (Bant Genişliği: 0.5)")
plt.show()

sns.kdeplot(data=tips, x="total_bill", bw_adjust=2)
plt.xlabel("Toplam Hesap")
plt.ylabel("Yoğunluk")
plt.title("Toplam Hesap Yoğunluğu (Bant Genişliği: 2)")
plt.show()
```

Bu örneklerde:

bw_adjust: Bant genişliğini ayarlar. Küçük değerler daha detaylı, büyük değerler daha pürüzsüz eğri oluşturur.<br>

### Gruplar Arasında Karşılaştırma (Hue Kullanımı)

hue parametresi kullanılarak, farklı kategorileri renklendirerek karşılaştırma yapabiliriz:

```python
# KDE plot oluşturma, cinsiyete göre renklendirme
sns.kdeplot(data=tips, x="total_bill", hue="sex")
plt.xlabel("Toplam Hesap")
plt.ylabel("Yoğunluk")
plt.title("Toplam Hesap Yoğunluğu (Cinsiyete Göre)")
plt.show()

#Bu örnekte:

# hue="sex": Veriyi cinsiyete göre renklendirir.
```

Bu grafik, bir Kernel Density Estimation (KDE) plot olup, body_mass_g değişkeninin dağılımını göstermektedir. Grafik, penguenlerin vücut kütlelerinin gram cinsinden yoğunluğunu görselleştirir.<br>

Grafiğin Açıklaması
X Ekseni (body_mass_g): Yatay eksen, penguenlerin vücut kütlesini gram cinsinden gösterir. Değerler yaklaşık 2000 gram ile 7000 gram arasında değişmektedir.<br>

Y Ekseni (Density): Dikey eksen, yoğunluğu temsil eder. Bu eksen, belirli bir vücut kütlesi aralığında penguenlerin yoğunluğunu gösterir.<br>

Yoğunluk Eğrisi: Grafik üzerindeki eğri, KDE eğrisidir ve penguenlerin vücut kütlesinin olasılık yoğunluğunu gösterir.<br>

Grafikten Çıkarımlar<br>

Yoğunluk Zirvesi: Grafik, yaklaşık 3750 gram civarında bir yoğunluk zirvesi göstermektedir. Bu, penguenlerin çoğunun vücut kütlesinin bu değere yakın olduğunu gösterir.<br>

Çift Tepe (Bimodal Dağılım): Grafik, iki belirgin tepe noktası (zirve) göstermektedir. İlk tepe yaklaşık 3750 gramda, ikinci tepe ise yaklaşık 5000 gramda. Bu, veri setinin bimodal (çift tepe) dağılım gösterdiğini ve iki farklı yoğunluk bölgesinin olduğunu gösterir.<br>

Dağılım Aralığı: Penguenlerin vücut kütleleri 2000 gram ile 7000 gram arasında değişmektedir, ancak yoğunluk büyük ölçüde 3000 ile 6000 gram arasında yoğunlaşmıştır.<br>

Yoğunluk Düşüşü: Grafik, 6000 gramın üzerinde ve 3000 gramın altında yoğunluğun azaldığını göstermektedir. Bu, çok az sayıda penguenin bu kütle aralıklarında olduğunu gösterir.<br>
