**Party Size(Parti Büyüklüğü): Bu sütun, daha büyük partilerin daha yüksek toplam fatura ve bahşiş tutarlarına sahip olma eğiliminde olup olmadığını görmek için kullanılabilir.

Dağılım grafiği, her bir işlem için toplam fatura tutarı (total_bill) ile bahşiş tutarı (tip) arasındaki ilişkiyi göstermektedir.<br>

Axes:<br>

X ekseni dolar cinsinden toplam fatura tutarını temsil eder.<br>
Y ekseni bahşiş miktarını dolar cinsinden temsil eder.<br>

Pozitif Korelasyon:<br>
Grafik, toplam hesap ile bahşiş arasında pozitif bir korelasyon olduğunu göstermektedir. Bu, toplam fatura tutarı arttıkça bahşiş tutarının da artma eğiliminde olduğu anlamına gelir.<br>

Veri noktalarının çoğu yukarı doğru bir eğilim izlemektedir, bu da insanların toplam hesapları daha yüksek olduğunda genellikle daha fazla bahşiş verdiklerini göstermektedir.<br>

Cluster of Data Points(Veri Noktaları Kümesi):<br>

Toplam fatura tutarları 10 ila 30 $ arasında büyük bir veri noktası kümesi görülmektedir. Buna paralel olarak, bu aralıktaki bahşiş miktarları çoğunlukla 1 ila 5 dolar arasındadır.<br>
Bu da çoğu işlemin hem toplam fatura hem de bahşiş için bu aralıkta yer aldığını göstermektedir.<br>

Variability(Değişkenlik):<br>

Daha yüksek faturalarda daha yüksek bahşişlere ilişkin genel bir eğilim olsa da, veri noktaları değişkenlik göstermektedir. <br>
Örneğin, bazı müşteriler benzer fatura tutarları için farklı bahşişler verebilir.<br>
Aynı fatura tutarı için verilen bahşişlerin önemli ölçüde farklılık gösterdiği birkaç nokta vardır, bu da farklı bahşiş davranışlarına işaret etmektedir.<br>

Outliers(Aykiri Degerler)<br>

Çizimde birkaç aykırı nokta var. Örneğin, bir nokta yaklaşık 10 $ bahşiş ile yaklaşık 50 $'lık bir toplam faturayı göstermektedir. Bu aykırı değerler daha büyük partileri, istisnai hizmeti veya diğer faktörleri temsil ediyor olabilir.<br>


**Summary(Ozet)** 

Özetle, dağılım grafiği toplam fatura ile bahşiş miktarı arasında pozitif bir korelasyon olduğunu göstermektedir, yani daha yüksek faturalar genellikle daha yüksek bahşişlere yol açmaktadır. Bununla birlikte, bahşiş davranışında değişkenlik vardır ve bahşiş miktarlarının genel eğilimden önemli ölçüde farklı olduğu bazı aykırı değerler mevcuttur.

```python
fig, ax = plt.subplots()

ax.scatter(tips.total_bill, tips.tip)

x=tips.total_bill
y=tips.tip

m, b = np.polyfit(x, y, 1) #Basit doğrusal regresyon hesaplaması

ax.plot(x, m*x + b, c="r")

plt.show()
```

Yeni dağılım grafiği, toplam fatura (total_bill) ile bahşiş (tip) arasındaki ilişkinin daha ayrıntılı bir analizini sağlayan bir regresyon çizgisi içerir.<br>

Plot Elements:<br>

1-Scatter Plot: Mavi noktalar, x ekseninde toplam fatura tutarını ve y ekseninde bahşiş tutarını gösteren bireysel veri noktalarını temsil etmektedir.<br>

2- Regression Line:<br>

Kırmızı çizgi, doğrusal regresyon kullanılarak hesaplanan en iyi uyum çizgisidir (regresyon çizgisi). Bu çizgi, toplam fatura ile bahşiş arasında öngörülen ilişkiyi temsil etmektedir.<br>

Yorumlama:

1-Positive Linear Relationship(Pozitif Doğrusal İlişki):

Kırmızı çizginin yukarı doğru eğimi, toplam hesap ile bahşiş arasında pozitif doğrusal bir ilişki olduğunu doğrulamaktadır. Bu da ortalama olarak toplam hesap arttıkça bahşiş miktarının da arttığı anlamına gelmektedir.<br>

2- Slope and Intercept(Eğim ve Kesişim):

Regresyon doğrusunun denklemi şu şekilde yazılabilir: tip =m x total_bill + b, m egim  b ise ara degerdir.

Kodda, np.polyfit(x, y, 1) bu değerleri hesaplar ve kırmızı çizgiyi çizmek için kullanılır. Eğim (m), toplam fatura ile ucun artma oranını gösterir.<br.

3-Data Fit(Veri Uyumu):

Kırmızı çizgi verilerin genel eğilimini göstermektedir, ancak tüm noktalar tam olarak çizgi üzerinde yer almamaktadır. Bu, toplam faturanın bahşişin belirlenmesinde önemli bir faktör olmasına rağmen, değişkenliğe neden olan başka faktörlerin de rol oynadığını göstermektedir.<br>
Veri noktalarının çizgi etrafında dağılması, benzer hesap tutarları için bile bahşiş verme davranışında değişkenlik olduğunu göstermektedir.<br>

4-Outliers and Variability(Aykırı Değerler ve Değişkenlik):<br>

Bazı noktalar regresyon çizgisinden uzakta yer almaktadır, bu da verilen bahşişin doğrusal modelin öngördüğünden çok daha yüksek veya çok daha düşük olduğu aykırı değerleri göstermektedir.<br>
Noktaların çizgi etrafında yayılması, doğrusal modelin genel eğilimi yakalamasına rağmen, bireysel ipuçlarını mükemmel bir şekilde tahmin etmediğini göstermektedir.<br>

Summary(Özet):
Regresyon doğrusu ile birlikte dağılım grafiği, daha yüksek toplam faturalarla birlikte bahşişlerin artmasına yönelik genel bir eğilim olduğunu göstermektedir. Ancak, regresyon çizgisi etrafındaki veri noktalarının dağılımından da anlaşılacağı üzere bahşiş verme davranışında değişkenlik söz konusudur. Bu durum, toplam fatura tutarının bahşişin belirlenmesinde önemli bir faktör olmakla birlikte, diğer faktörlerin de bahşiş tutarını etkileyebileceğini göstermektedir.

Çubuk grafik, erkek ve kadın müşteriler tarafından verilen bahşişlerin karşılaştırmasını göstermektedir.<br>

Plot Elements:

X-axis: X ekseni, iki kategori ile müşterilerin cinsiyetini temsil eder: Kadın ve Erkek.<br>

Y-axis: Y ekseni verilen toplam bahşiş miktarını temsil etmektedir.<br>

### Interpretation

1-Comparison of Tips(İpuçlarının Karşılaştırılması):

Çubukların yüksekliği her bir cinsiyet tarafından verilen toplam bahşiş miktarını göstermektedir.<br>
Erkek çubuğunun Kadın çubuğundan daha yüksek olması, erkek müşterilerin ortalama olarak kadın müşterilere kıyasla daha yüksek bahşiş verdiğini göstermektedir.<br>

2-Numerical Difference(Sayısal Fark):

Kesin rakamlar gösterilmese de, erkek müşteriler tarafından verilen toplam bahşişlerin kadın müşteriler tarafından verilenlerden önemli ölçüde daha yüksek olduğu açıktır.<br>
Bunun nedeni erkeklerin daha fazla harcama yapması, daha fazla sayıda erkek müşteri olması veya farklı bahşiş verme davranışları gibi çeşitli nedenler olabilir.<br>


### Possible Explanations and Further Analysis(Olası Açıklamalar ve İleri Analizler):<br>

1-Spending Habits(Harcama Alışkanlıkları):<br>
Erkek müşterilerin faturalarında daha fazla harcama yapıyor olması ve bunun da daha yüksek bahşişlere yol açması mümkündür.<br>

2-Sample Size(Örnek Büyüklüğü):
Veri kümesinde kadın müşterilerden daha fazla erkek müşteri olabilir ve bu da erkekler için toplam bahşiş miktarının daha yüksek olmasına yol açabilir.<br>

3-Tipping Behavior(Bahşiş Verme Davranışı):<br>

Bahşiş verme davranışında cinsiyetler arasında bir fark olabilir; erkekler muhtemelen faturanın daha yüksek bir yüzdesini bahşiş olarak vermektedir.<br>


### Summary:

Çubuk grafik, erkek müşterilerin kadın müşterilere kıyasla daha fazla bahşiş verdiğini göstermektedir. Bunun nedeni daha yüksek harcama, daha fazla sayıda erkek müşteri veya farklı bahşiş alışkanlıkları olabilir. Bu farkın arkasındaki nedenleri anlamak için, toplam fatura tutarları ve cinsiyet başına işlem sayısı gibi diğer faktörleri dikkate alan daha fazla analiz yararlı olacaktır.

```python
average_tips = tips.groupby('sex')['tip'].mean()

# Çubuk grafiğini çizme
fig, ax = plt.subplots()

ax.bar(average_tips.index, average_tips.values)

ax.set_xlabel("Sex")  # cinsiyet sutununa gore gruplar
ax.set_ylabel("Average Tip") # 
ax.set_title("Average Tip by Sex")

plt.show()
```

```python
average_tips = tips.groupby('sex')['tip'].sum()

# Çubuk grafiğini çizme
fig, ax = plt.subplots()

ax.bar(average_tips.index, average_tips.values)

ax.set_xlabel("Sex")
ax.set_ylabel("Total Tip")
ax.set_title("Total Tip by Sex")

plt.show()
```

```python
z = np.array([16, 21, 45, 19])

mylabels = ["Apples", "Bananas", "Cherries", "Melon"]

plt.figure(figsize=(10, 8))

myexplode = [0.2, 0, 0, 0]

plt.pie(z, labels = mylabels, 
        labeldistance = 0.7, 
        explode = myexplode, 
        autopct = "%.2f",
       startangle = 90)

plt.show()
```

```python
plt.figure(figsize=(10, 8))

color_list = ["gold", "yellowgreen", "lightcoral", "lightskyblue"]
explode_list = [0.5, 0, 0, 0]

y = np.array([25, 35, 25, 15])

mylables = ["Apples", "Bananas", "Cherries", "Melon"] # isimler 

plt.pie(y, labels = mylables,
       labeldistance = 0.5, # yaziyi kenara ortaya ayarlar
       autopct = "%.1f",
       startangle = 90,
       shadow = True,  # golgelendirme 
       pctdistance=0.3,
       colors = color_list,
       explode = explode_list)

plt.legend(loc = (1))
plt.show()
```

```python
plt.figure(figsize=(10, 8))

sex_counts = tips['sex'].value_counts() # 'sex' sütununa göre veri sayısını hesaplama

color_list = ["lightskyblue", "lightcoral"]
explode_list = [0.1, 0]


plt.pie(sex_counts, labels=sex_counts.index, 
        autopct='%1.1f%%', 
        startangle=140, 
        pctdistance=0.5,
        labeldistance = 0.4,
        shadow = True,
        colors = color_list,
       explode = explode_list)
plt.title("Tips Dataset - Meal Counts by Sex")

plt.show()
```
